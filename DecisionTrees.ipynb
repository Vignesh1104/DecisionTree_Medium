```
from copy import deepcopy
def gini(daf,c):

  '''This Function calculates the gini index  for each features of the 
  dataframe except for the features speccified in the c parameter'''
  
  col=deepcopy(c)  # deepcopy - any changes made in the duplicate list  will not affect the original list
  val=1
  values=daf[col].unique()
  x=(daf.groupby(col))['Eligibility'].agg('count')
  for i in x:
    val-=(i/sum(x))**2
  return val,values
```

```
def cover(df,x):

  '''This Function splits the parent node into left and right node
   by calling the gini function''' 
  
  ex=deepcopy(x)  # deepcopy - any changes made in the duplicate list  will not affect the original list
  dic=[]
  cols=list(df.columns[:-1])
  if len(ex)>0:
    for splitted in ex: 
      cols.remove(splitted)
  for i in cols:
    z,values=gini(df,i)
    dic.append((i,z,values))
  dic.sort(key=lambda x:x[1])
  ref=dic[0]
  ex.append(ref[0])
  l=new_daf=df[df[ref[0]]==ref[2][0]]
  r=new_daf=df[df[ref[0]]==ref[2][1]] 
  return l,r,ex
```

```
def split(df,p_c):
  '''This function is called recursively to 
  split the dataset ny calling the cover function '''
  processed_cols=deepcopy(p_c)
  l,r,c=cover(df,processed_cols)
  if len(l['Eligibility'].unique())>1:
    split(l,c)
  else:
    print(l)
    print('\n')  
  if len(r['Eligibility'].unique())>1:
    split(r,c)
  else:
    print(r)
    print('\n')
lst=[]    
print('The Leaves/ Terminal Nodes are :\n')    
split(df,lst)
```
